### Tidying the data

Next, you will further tidy the two dataframes **stars** and **genres**.

Let's start with **stars**.

```python
# Transform the dataframe "stars" to a new dataframe named "stars_long" with the following four columns:
# Series_Title, Released_Year, StarNo (e.g., Star1, Star2, ...), StarName
# Hint: use melt()
stars_long = stars.melt(id_vars=['Series_Title', 'Released_Year'], value_vars=['Star1', 'Star2', 'Star3', 'Star4'], var_name='StarNo', value_name='StarName')
stars_long
```

```python
# Can you transform dataframe 'stars_long' back to its original shape?
# Hint: use pivot()
stars_long.pivot(index=['Series_Title', 'Released_Year'], columns='StarNo', values='StarName')
```

```python
# In dataframe "stars_long", find all movies that star "Morgan Freeman".
stars_long[stars_long.StarName == 'Morgan Freeman']
```

```python
# Who appeared in Star2 the most times? List the top five actors.
stars_long[stars_long['StarNo'] == 'Star2']['StarName'].value_counts().head()
```

```python
# Who stars in the most movies in this list? List the top 20 actors.
stars_long.StarName.value_counts().head()
```

```python
# Which movie stars had the highest total gross in this movie list? Show the top 10 actors.
# Hint: Join "stars_long" and "imdb"; then group by StarName
merged_df = pd.merge(stars_long, imdb[['Series_Title', 'Released_Year', 'Gross']], on=['Series_Title', 'Released_Year'], how='left')
total_gross_by_star = merged_df.groupby('StarName')['Gross'].sum()
top_10_stars = total_gross_by_star.sort_values(ascending=False).head(10)
display(top_10_stars)
```

```python
# Find the best director-actor duos
# i.e., director-actor pairs that collaborated in at least five movies,
# sort them in descending order of average IMDB_Rating
# Hint: Join imdb and stars_long; group by ['Director','StarName']

duos_df = pd.merge(imdb[['Series_Title', 'Released_Year', 'Director', 'IMDB_Rating']],
    stars_long[['Series_Title', 'Released_Year', 'StarName']],
    on=['Series_Title', 'Released_Year'],
    how='inner'
)

duo_stats = duos_df.groupby(['Director', 'StarName']).agg(
    collaboration_count=('Series_Title', 'count'),
    average_imdb_rating=('IMDB_Rating', 'mean')
).reset_index()

best_duos = duo_stats[duo_stats['collaboration_count'] >= 5]

best_duos = best_duos.sort_values(by='average_imdb_rating', ascending=False)

display(best_duos)
```

Next, let's reshape the dataframe **genres**, which is a little bit more complicated.

```python
# Step 1: Split the 'Genre' string by ', ' into a list of individual genres and expand them to different columns
genres_split = genres['Genre'].str.split(', ', expand=True)
genres_split
```

```python
# Step 2: Rename the columns as: Genre1, Genre2, Genre3
genres_split.columns = ['Genre1', 'Genre2', 'Genre3']
genres_split
```

```python
# Step 3: Combine ['Series_Title','Released_Year'] in 'genres' and ['Genre1','Genre2','Genre3'] in 'genres_split'.
# Save it to a new dataframe named 'genres_wide'.
# Hint: Use pd.concat(...)
genres_wide = pd.concat([genres[['Series_Title','Released_Year']], genres_split], axis=1)
genres_wide
```

```python
# Step 4: Transform genres_wide to a new dataframe genres_long with the following four columns:
# Series_Title, Released_Year, GenreNo (e.g., Genre1, Genre2, Genre3), GenreName
# Hint: use melt()
genres_long = genres_wide.melt(id_vars=['Series_Title', 'Released_Year'], value_vars=['Genre1', 'Genre2', 'Genre3'], var_name='GenreNo', value_name='GenreName')
genres_long = genres_long.dropna(subset=['GenreName'])
genres_long
```

```python
# How many unique genres (atomic values, e.g., Drama, Comedy, ...) are there?
# How many movies are there for each genre?
unique_genres_count = genres_long['GenreName'].nunique()
genres_movies_count = genres_long['GenreName'].value_counts()

print(f"Number of unique genres: {unique_genres_count}")
print("\nNumber of movies per genre:")
display(genres_movies_count)
```

```python
# What is the average IMDB rating for each genre?
# Sort the genres in descending order of average IMDB_Rating.
# Hint: join imdb with genres_long; group by GenreName

genre_imdb_df = pd.merge(genres_long, imdb[['Series_Title', 'Released_Year', 'IMDB_Rating']], on=['Series_Title', 'Released_Year'], how='inner')
average_imdb_rating_per_genre = genre_imdb_df.groupby('GenreName')['IMDB_Rating'].mean().sort_values(ascending=False)
display(average_imdb_rating_per_genre)
```

```python
# Who is the "King of Comedy" (i.e., the actor who starred in the most comedy movies)?
# Hint: find all comedies in genres_long ; join with stars_long; group by StarName

comedy_movies = genres_long[genres_long['GenreName'] == 'Comedy']
comedy_stars = pd.merge(comedy_movies, stars_long, on=['Series_Title', 'Released_Year'], how='inner')
king_of_comedy = comedy_stars['StarName'].value_counts().head(1)
display(king_of_comedy)
```

```python
# Who are the best action stars?
# i.e., the actors who star in at least 5 action movies, sorted by average IMDB rating in descending order
# Hint: You need to join columns from all three dataframes: imdb, stars_long, genres_long

action_movies = genres_long[genres_long['GenreName'] == 'Action']

action_stars_df = pd.merge(action_movies, stars_long, on=['Series_Title', 'Released_Year'], how='inner')

action_stars_with_ratings = pd.merge(
    action_stars_df,
    imdb[['Series_Title', 'Released_Year', 'IMDB_Rating']],
    on=['Series_Title', 'Released_Year'],
    how='inner'
)

star_action_stats = action_stars_with_ratings.groupby('StarName').agg(
    num_action_movies=('Series_Title', 'count'),
    average_imdb_rating=('IMDB_Rating', 'mean')
).reset_index()

best_action_stars = star_action_stats[star_action_stats['num_action_movies'] >= 5].sort_values(
    by='average_imdb_rating', ascending=False
)
display(best_action_stars)
```

```python
# Create a pivot table to show the number of movies of different genres in different decades.
# Row: GenreName
# Column: Decade
# Hint: https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html

genres_decade_df = pd.merge(
    genres_long,
    imdb[['Series_Title', 'Released_Year', 'Decade']],
    on=['Series_Title', 'Released_Year'],
    how='inner'
)

pivot_table_genres_decade = pd.pivot_table(
    genres_decade_df,
    index='GenreName',
    columns='Decade',
    aggfunc='count',
    values='Series_Title',
    fill_value=0
)
display(pivot_table_genres_decade)
```
